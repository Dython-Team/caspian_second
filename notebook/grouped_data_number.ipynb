```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.table import Table
```

```python
df = pd.read_csv('../final_data/merged_dataset.csv')
```

```python
df['Source'].unique()
```

```python
df1 = df[df['Source']=='casp1']
df3 = df[df['Source']=='casp3']
df4 = df[df['Source']=='casp4']
df5 = df[df['Source']=='casp5']
```

```python
grouped = df1.groupby(['sex', 'age']).size().reset_index(name='count')
# Pivoting the data for better visualization in the table
table_data = grouped.pivot(index='age', columns='sex', values='count').fillna(0).astype(int)
fig, ax = plt.subplots(figsize=(8, 6))
ax.axis('off')  # Hide axes
ax.set_title("Sample Count by Age and Sex - casp1", fontsize=16, weight='bold')
table = Table(ax, bbox=[0, 0, 1, 1])
def add_cell(row, col, text, loc='center', facecolor='white'):
    table.add_cell(
        row, col, text=text, loc=loc, width=0.25, height=0.075, facecolor=facecolor
    )

# Add header cells
for col, sex in enumerate(['Age'] + list(table_data.columns), start=0):
    add_cell(0, col, sex, facecolor='#40466e', loc='center')

# Add data cells
for row_idx, (age, row) in enumerate(table_data.iterrows(), start=1):
    add_cell(row_idx, 0, age, facecolor='#dbe4f0', loc='center')  # Age
    for col_idx, value in enumerate(row, start=1):
        add_cell(row_idx, col_idx, value, facecolor='white', loc='center')

# Add the table to the plot
ax.add_table(table)
plt.savefig('sample_count_casp1.png', dpi=300, bbox_inches='tight')
plt.show()
```
